```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score

###############################################################################
# generate some sparse data to play with
np.random.seed(42)

n_samples, n_features = 50, 200
X = np.random.randn(n_samples, n_features)
coef = 3 * np.random.randn(n_features)
inds = np.arange(n_features)
np.random.shuffle(inds)
coef[inds[10:]] = 0  # sparsify coef
y = np.dot(X, coef)

# add noise
y += 0.01 * np.random.normal((n_samples,))

# Split data in train set and test set
n_samples = X.shape[0]
X_train, y_train = X[:n_samples / 2], y[:n_samples / 2]
X_test, y_test = X[n_samples / 2:], y[n_samples / 2:]

###############################################################################
# Lasso
from sklearn.linear_model import Lasso

alpha = 0.1
lasso = Lasso(alpha=alpha)

y_pred_lasso = lasso.fit(X_train, y_train).predict(X_test)
r2_score_lasso = r2_score(y_test, y_pred_lasso)
print(lasso)
print("r^2 on test data : %f" % r2_score_lasso)

###############################################################################
# ElasticNet
from sklearn.linear_model import ElasticNet

enet = ElasticNet(alpha=alpha, l1_ratio=0.7)

y_pred_enet = enet.fit(X_train, y_train).predict(X_test)
r2_score_enet = r2_score(y_test, y_pred_enet)
print(enet)
print("r^2 on test data : %f" % r2_score_enet)

plt.plot(enet.coef_, label='Elastic net coefficients')
plt.plot(lasso.coef_, label='Lasso coefficients')
plt.plot(coef, '--', label='original coefficients')
plt.legend(loc='best')
plt.title("Lasso R^2: %f, Elastic Net R^2: %f"
          % (r2_score_lasso, r2_score_enet))
plt.show()
```

```python
import re
import nltk

#parser = nltk.parse.malt.MaltParser()

stopwords = nltk.corpus.stopwords.words('english')


def tokenize(text):
    tokens_words = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent) if
              word.lower() not in stopwords]
    tokens_sentences=[sent for sent in nltk.sent_tokenize(text)]
    filtered_tokens = []
    for token in tokens_words:
        if re.search('[a-zA-Z]', token):
            token = token.replace('*', '')
            token = token.replace('[Delayed]', '')
            filtered_tokens.append(token)

    return filtered_tokens, tokens_sentences

grammar = r"""
    NP : {<DT|PP\$>?<JJ>*<NN>}   # determiner/possessive, adjective, noun
         {<NNP>+}                # sequences of proper nouns
         {<NN>+}                 # sequence of common nouns
         }<VBD|IN>+{             # Chink sequences of VBD and IN
  """
cp = nltk.RegexpParser(grammar)


with open('news_first_100.txt') as f:
    with open('tokens_first_100.txt', 'w') as t:
        for each in f.readlines():
            print "\n"
            print each[:-1]
            each = each[each.find(';'):]
            tokens, sentences = tokenize(each)
            print nltk.pos_tag(tokens)
            print cp.parse(nltk.pos_tag(tokens))
            #print parser.parse(nltk.pos_tag(tokens))
            
            print tokens
            print sentences
            t.writelines(" ".join(tokens) + "\n")
```
